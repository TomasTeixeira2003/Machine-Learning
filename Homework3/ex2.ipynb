```python
import numpy as np
```

```python
W_1 = [[1, 1, 1, 1],
       [1, 1, 2, 1],
       [1, 1, 1, 1]]
b_1 = np.ones((3, 1))

W_2 = [[1, 4, 1],
       [1, 1, 1]]
b_2 = np.ones((2, 1))

W_3 = [[1, 1],
       [3, 1],
       [1, 1]]
b_3 = np.ones((3, 1))

X_01 = np.ones((4, 1))

X_02 = [[1], [0], [0], [-1]]

Z_B = [[-1], [1], [-1]]

Z_A = [[1], [-1], [-1]]
```

```python
def tanh(x):
    return (np.exp(0.5*x - 2) - np.exp(-0.5*x +2))/(np.exp(0.5*x - 2) + np.exp(-0.5*x +2))

def tanh_grad(x):
    return 0.5 * (1-tanh(x)**2)

def squared_error(z, x):
    return 0.5 * (z-x)**2

def squared_error_grad(z, x):
    return x-z
```

```python
Z_11 = np.dot(W_1, X_01) + b_1
Z_12 = np.dot(W_1, X_02) + b_1

X_11 = tanh(Z_11)
X_12 = tanh(Z_12)

print("Z_11:\n", Z_11)
print("Z_21:\n", Z_12)
print("X_11:\n", X_11)
print("X_21:\n", X_12)
```

```python
Z_21 = np.dot(W_2, X_11) + b_2
Z_22 = np.dot(W_2, X_12) + b_2

X_21 = tanh(Z_21)
X_22 = tanh(Z_22)

print("Z_22:\n", Z_21)
print("Z_22:\n", Z_22)
print("X_21:\n", X_21)
print("X_22:\n", X_22)
```

```python
Z_31 = np.dot(W_3, X_21) + b_3
Z_32 = np.dot(W_3, X_22) + b_3

X_31 = tanh(Z_31)
X_32 = tanh(Z_32)

print("Z_31:\n", Z_31)
print("Z_32:\n", Z_32)
print("X_31:\n", X_31)
print("X_32:\n", X_32)
```

```python
delta_31 = squared_error_grad(Z_B, X_31)*tanh_grad(Z_31)
delta_32 = squared_error_grad(Z_A, X_32)*tanh_grad(Z_32)

W_31_grad = (squared_error_grad(Z_B, X_31)*tanh_grad(Z_31)).dot(X_21.T)
W_32_grad = (squared_error_grad(Z_A, X_32)*tanh_grad(Z_32)).dot(X_22.T)


print("delta_31:\n", delta_31)
print("delta_32:\n", delta_32)
print("W_31_grad:\n", W_31_grad)
print("W_32_grad:\n", W_32_grad)
print(squared_error_grad(Z_B, X_31))
```

```python
delta_21 = ((np.array(W_3).T).dot(delta_31))*tanh_grad(Z_21)
delta_22 = ((np.array(W_3).T).dot(delta_32))*tanh_grad(Z_22)

W_21_grad = (((np.array(W_3).T).dot(delta_31))*tanh_grad(Z_21)).dot(X_11.T)
W_22_grad = (((np.array(W_3).T).dot(delta_32))*tanh_grad(Z_22)).dot(X_12.T)


print("delta_21:\n", delta_21)
print("delta_22:\n", delta_22)
print("W_21_grad:\n", W_21_grad)
print("W_22_grad:\n", W_22_grad)
```

```python
delta_11 = ((np.array(W_2).T).dot(delta_21))*tanh_grad(Z_11)
delta_12 = ((np.array(W_2).T).dot(delta_22))*tanh_grad(Z_12)

W_11_grad = (((np.array(W_2).T).dot(delta_21))*tanh_grad(Z_11)).dot(X_01.T)
W_12_grad = (((np.array(W_2).T).dot(delta_22))*tanh_grad(Z_12)).dot(np.array(X_02).T)

print("delta_11:\n", delta_11)
print("delta_12:\n", delta_12)
print("W_11_grad:\n", W_11_grad)
print("W_12_grad:\n", W_12_grad)
```

```python
W_1_grad_total = W_11_grad + W_12_grad
W_2_grad_total = W_21_grad + W_22_grad
W_3_grad_total = W_31_grad + W_32_grad
```

```python
lr = 0.1

W_1_grad_total = W_11_grad + W_12_grad

W_1 = W_1 - lr * W_1_grad_total

print("new W_1:\n", W_1)
```

```python
lr = 0.1

W_2_grad_total = W_21_grad + W_22_grad

W_2 = W_2 - lr * W_2_grad_total

print("new W_2:\n", W_2)
```

```python
lr = 0.1

W_3_grad_total = W_31_grad + W_32_grad

W_3 = W_3 - lr * W_3_grad_total

print("new W_3:\n", W_3)
```
